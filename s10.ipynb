# Confidence Interval

```python
import pandas as pd 
import numpy as np
df = pd.read_csv('heart.csv')
df
```

```python
df['sex1'] = df.sex.replace({1:'Male', 0:'Female'}) #mengganti menjadi data hot encoding
```

```python
dx = df[['target', 'sex1']].dropna() #menghapus nilai nan
```

```python
pd.crosstab(dx.target, dx.sex1) #mengambil kolom jumlah penyakit jantung untuk laki-laki dan perempuan
```

```python
p_fm = 226/(86+226)
p_fm
```

```python
n = 86+226 
```

```python
se_female = np.sqrt(p_fm*(1-p_fm)/n)
se_female
```

```python
#nilai z-score adalah 1.96 untuk CI 95% 
z_score = 1.96
lcb = p_fm - z_score * se_female #batas bawah CI
ucb = p_fm + z_score * se_female #batas atas CI
lcb, ucb
```

```python
import statsmodels.api as sm
sm.stats.proportion_confint(n * p_fm, n )
```

# Hypothesis Testing

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/nhanes_2015_2016.csv')
df.head()
```

H0 = rata-rata BMI laki-laki sama dengan rata-rata BMI perempuan\
H1 = rata-rata BMI laki-laki berbeda rata-rata BMI perempuan

```python
female = df[df['RIAGENDR']==2]
male = df[df['RIAGENDR']==1]
```

```python
n1 = len(female)
mu1 = female['BMXBMI'].mean()
sd1 = female['BMXBMI'].std()
(n1, mu1, sd1)
```

```python
sm.stats.ztest(female['BMXBMI'].dropna(), male['BMXBMI'].dropna(), alternative='two-sided') #'two-sided' untuk memriksa ketidaksamaan
```
