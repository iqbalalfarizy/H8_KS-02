# Basic Visualization

```python
import pandas as pd
import numpy as np
```

```python
df_can = pd.read_excel('https://github.com/ardhiraka/PFDS_sources/blob/master/Canada.xlsx?raw=true',
sheet_name='Canada by Citizenship',
skiprows=range(20),
skipfooter=2)

df_can.head()
```

```python
df_can.info()
```

```python
df_can.index
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
df_can.head()
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent','RegName':'Region'}, inplace=True)
df_can.columns
```

```python
df_can['Total']=df_can.sum(axis=1)
```

```python
df_can.describe()
```

## Line Plot

```python
haiti = df_can.loc['Haiti', years] #years(1980-2013)
haiti.plot()
```

```python
haiti.index = haiti.index.map(int)

haiti.plot(kind='line')
plt.title('Imigrasi dari Haiti')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.show()
```

```python
haiti.plot(kind='line')
plt.title('Imigrasi dari Haiti')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.text(2000, 6000, 'Gempa bumi 2010')
plt.show()
```

```python
china_india = df_can.loc[['China', 'India'], years]
china_india.head()
```

```python
china_india.plot(kind='line')
```

```python
df_CI = china_india.transpose()
df_CI.head()
```

```python
df_CI.plot(kind='line')
plt.title('Imigrasi dari China dan India')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.show()
```

```python
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = df_can.head()
df_top5 = df_top5[years].transpose()
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line', figsize = (20,12))
plt.title('Imigrasi 5 negara terbanyak')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.show()
```

## Area Plot

```python
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = df_can.head()
df_top5 = df_top5[years].transpose()
df_top5.index = df_top5.index.map(int)

df_top5.plot(kind='area',
            stacked=False,
            figsize = (20,12))
plt.title('Imigrasi 5 negara terbanyak')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.show()
```

# Histogram

```python
df_can['2013'].head()
```

```python
import numpy as np
count, bin_edges = np.histogram(df_can['2013']) # default 10 bin

print(count) # 178 negara dst
print(bin_edges)# 0 sampai 3412.9 imigran dst
```

```python
df_can['2013'].plot(kind='hist', figsize=(8,5))
plt.title('Imigran 195 negara tahun 2013')
plt.xlabel('jumlah imigran')
plt.ylabel('jumlah negara')
plt.show()
```

```python
df_can['2013'].plot(kind='hist', figsize=(8,5), xticks=bin_edges) #xticks untuk menyesuaikan data dengan bin_edges
plt.title('Imigran 195 negara tahun 2013')
plt.xlabel('jumlah imigran')
plt.ylabel('jumlah negara')
plt.show()
```

```python
df_t = df_can.loc[['Indonesia', 'Malaysia', 'Singapore'], years].transpose()
df_t.head()
```

```python
df_t.plot(kind='hist', figsize=(10,10))
plt.title('Imigran negara Indonesia, Malaysia, dan Singapura')
plt.xlabel('jumlah imigran')
plt.ylabel('jumlah tahun')
plt.show()
```

```python
count, bin_edges = np.histogram(df_t, 20) #nilai xticks
print(count)
print(bin_edges)

df_t.plot(kind='hist',
        figsize=(15,10),
        bins=20,
        alpha=0.5,
        color = ['coral', 'darkslateblue', 'mediumseagreen'],
        xticks=bin_edges) #xticks untuk menyesuaikan data dengan bin_edges
plt.title('Imigran negara Indonesia, Malaysia, dan Singapura')
plt.xlabel('jumlah imigran')
plt.ylabel('jumlah tahun')
plt.show()
```

```python
#list ketersediaan warna
import matplotlib
for name, hex in matplotlib.colors.cnames.items():
    print(name, hex)
```

```python
count, bin_edges = np.histogram(df_t, 20)
xmin = bin_edges[0]-10
xmax = bin_edges[-1]+10

print(count)
print(bin_edges)

df_t.plot(kind='hist',
        figsize=(15,10),
        bins=20,
        alpha=0.6,
        color = ['coral', 'darkslateblue', 'mediumseagreen'],
        stacked=True, #plot tidak tumpang tindih
        xlim=(xmin, xmax), #menghilangkan celah ekstra pada tepi plot
        xticks=bin_edges) #xticks untuk menyesuaikan data dengan bin_edges
plt.title('Imigran negara Indonesia, Malaysia, dan Singapura')
plt.xlabel('jumlah imigran')
plt.ylabel('jumlah tahun')
plt.show()
```

## sub plot

```python
fig = plt.figure() #membuat figure
ax0 = fig.add_subplot(1,2,1) # 1 row, 2 column, first plot
ax1 = fig.add_subplot(1,2,2) # 1 row, 2 column, second plot

# subplot 1 : box plot
df_CI.plot(kind='box', color='blue', vert=False, figsize=(20,6), ax=ax0)
ax0.set_title('boxplot imigran dari cina dan india')
ax0.set_xlabel('jumlah imigran')
ax0.set_ylabel('negara')

# subplot 2 : line plot
df_CI.plot(kind='line',  figsize=(20,6), ax=ax1)
ax1.set_title('boxplot imigran dari cina dan india')
ax1.set_xlabel('jumlah imigran')
ax1.set_ylabel('negara')
```

### membuat boxplot dari 15 negara terbanyak berdasarkan dekade

```python
df_top15 = df_can.sort_values(['Total'], ascending=False, axis=0).head(15)
df_top15
```

```python
years_80s = list(map(str, range(1980,1990)))
years_90s = list(map(str, range(1990,2000)))
years_00s = list(map(str, range(2000,2010)))

df_80s = df_top15.loc[:, years_80s].sum(axis=1)
df_90s = df_top15.loc[:, years_90s].sum(axis=1)
df_00s = df_top15.loc[:, years_00s].sum(axis=1)

new_df = pd.DataFrame({'1980s':df_80s, '1990s':df_90s, '2000s':df_00s,})
new_df
```

```python
new_df.plot(kind='box', figsize=(10,6))
plt.title('Imigrasi 15 negara berdasarkan decade 80s, 90s, dan 00s')
plt.show()
```

## Scatter Plot

```python
df_tot = pd.DataFrame(df_can[years].sum(axis=0))
df_tot.head()
```

```python
df_tot.index = map(int, df_tot.index)
df_tot.reset_index(inplace=True)
```

```python
df_tot.columns = ['year', 'total']
df_tot.head()
```

```python
df_tot.plot(kind='scatter', x='year', y='total', figsize=(8,6))
plt.title('total imigrasi dari kanada')
plt.xlabel('tahun')
plt.ylabel('jumlah imigran')
plt.show
```
